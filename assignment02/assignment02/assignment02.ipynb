# import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function: $f(x) = \arcsin(x)$

```python
def myFunction(x):
    f = np.arcsin(x)
    return f
```

# define the derivative of my function:
$f'(x) = \frac{1}{\sqrt{1-x^2}}$

```python
def myDerivativeFunction(x):
    Df = np.divide(1, np.sqrt(1-np.power(x, 2)))
    return Df
```

# define the domain of the function:
$x = [ -1 : 1 : 0.1 ]$

```python
x = np.arange(-1, 1, 0.1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

# plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# define Taylor approximation:
$\hat{f(x)} = f(x_0) + f'(x_0)(x-x_0)$

```python
def myTaylorApprox(x, p):
    TA = myFunction(p) + myDerivativeFunction(x) * (x - p)
    return TA
```

# pick three points:
p1 = -0.5  
p2 = 0  
p3 = 0.5

```python
p1 = -0.5
p2 = 0
p3 = 0.5
```

# define the domain of each approximations:
x1 = [-1 : 0.1 : 0]  
x2 = [-0.5 : 0.1 : 0.5]  
x3 = [0 : 0.1 : 1]

```python
x1 = np.arange(-1, 0, 0.1)
x2 = np.arange(-0.5, 0.5, 0.1)
x3 = np.arange(0, 1, 0.1)
```

# compute Taylor Approximations

```python
TA1 = myTaylorApprox(x1, p1)
TA2 = myTaylorApprox(x2, p2)
TA3 = myTaylorApprox(x3, p3)
```

# plot the graphs for the function, its derivative, and its Taylor expansions

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.plot(x1, TA1, 'g', label="Taylor Approximation at -0.5")
plt.plot(x2, TA2, 'c', label="Taylor Approximation at 0")
plt.plot(x3, TA3, 'm', label="Taylor Approximation at 0.5")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
